```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv('iris_data.csv')
```

```python
print(iris.shape)
```

```python
print(iris.columns)
```

```python
#how many data points for each class are present 

iris['species'].value_counts()
#this is the balanced dataset as each of the species value is 50 , unbalanced dataset ex diabetic patient in 10000 patients
```

```python
iris.plot(kind='scatter', x='sepal_length' , y='sepal_width')
plt.grid()
plt.show()
```

```python
# now we are unable to differentiate between different species of the flower on the graph so seaborn gives us 
#easy way to color different group of data
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", size=6).map(plt.scatter , "sepal_length" , "sepal_width").add_legend()
plt.show()
#hue specifies by which parameter i should color my dataset
#size specifies width of the graph
```

by 2-d scatter plot of sepal_width, sepal_length we can only seperate setosa for other two species
