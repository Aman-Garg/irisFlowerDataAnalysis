## Histograms, PDF, CDF
- how to plot 1d scatter plot
- probablity density function(PDF)
- Cumulative distribution function(CDF)

```python
#1d sscatter plot of petal-length 
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

plt.plot(iris_setosa['petal_length'], np.zeros_like(iris_setosa['petal_length']),'ro', label = 'Setosa')
plt.plot(iris_versicolor['petal_length'], np.zeros_like(iris_versicolor['petal_length']), 'bo', label = 'Versicolor')
plt.plot(iris_virginica['petal_length'], np.zeros_like(iris_virginica['petal_length']), 'go', label = 'virginica')
plt.xlabel('Petal Length')
plt.legend()
plt.show()

#now from graph we cannot find how many points are lying between 1 and 2, 
#green point are overlaping blue point
#so we will divide the x axis into more intervals like 1 - 2 to 1, 1.2 , 1.4 , 1,6 , 1.8 , 2.0
#and we will use y axis a a count means how many points lies between 1.0 and 1.2, 
#by doing the above two steps we are basically making the histogram 
#so for 1-D plot histogram is way better option than 1-D scatter plot
```

```python
#code to plot hostogram 
#distplot stands for distribution plot
#below we have plotted the histogram of petal_length of setosa, versicolor and virginica
#dark mountain like lines here are called prob.density function we need to join the centers of the each bar to draw this line 
#PDF is basically called smoothed histogram we smooth histogram it with the help of kernel density fucntion

sns.FacetGrid(iris, hue="species" , size=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend()
plt.show()
```

# Cumulative Distribution Function (CDF)
- starts at zero and ends at 1 as it defines the probablity of the data 
- Basically to find things like how much percent of the setosa flower have petal length below 82% and all 

### How you construct CDF 
- suppose you want to calculate the CDF value at point 1.6 then you count the number of setosa flower having petal length below or equal to 1.6, then divide it by total number of setosa flower i.e we are just finding the probablity of the flower having less than 1.6 petal-length 
- or we can find the sum of all the bars of PDF which have petal-length less than 1.6 and this process is called cumulative sum 
- CDF value at a particular point the basically area under the curve of PDF till that point so if we differentiate our CDF we get PDF and vice versa. - try to use this approach to find CDF



```python
#plot CDF of Petal-Length 

counts , bin_edges = np.histogram(iris_setosa['petal_length'], bins = 10, density = True)

pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

#compute CDF
#cumsum stands for cumulative sum
#when you apply np.cumsum function on PDF you get your CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label='PDF')
plt.plot(bin_edges[1:], cdf, label='CDF')
plt.xlabel('Petal-Lenght')
plt.ylabel('Probablity')
plt.legend()

plt.show()
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins = 10, density = True )

pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label='PDF for Setosa')
plt.plot(bin_edges[1:] , cdf , label='CDF for Setosa')

#versicolor
counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins = 10, density = True )

pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label='PDF for Versicolor')
plt.plot(bin_edges[1:] , cdf , label='CDF for Versicolor')

#virgnica
counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins = 10, density = True )

pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:] , pdf , label='PDF for Virginica')
plt.plot(bin_edges[1:] , cdf , label='CDF for Virginica')

plt.xlabel('Petal-Length')
plt.ylabel('Probablity')
#plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),ncol=3, fancybox=True, shadow=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# As we can see form the graph that if the petal length is 5 then ther is 95% probablity it is versicolor 
# where two PDF intersect we take that value as the threshold value 
```

# Mean, Variance, Stdandard Deviation 

```python
# we are printing mean of petal length of each species of flower 
print('Means')
#it tellls us about the centeral tendency
print(np.mean(iris_setosa['petal_length']))
print(np.mean(iris_versicolor['petal_length']))
print(np.mean(iris_virginica['petal_length']))

#Mean have this problem because of one outlier it can increase drastically
print('\nMean with an outlier of setosa flower petal-length')
print(np.mean(np.append(iris_setosa['petal_length'] , 50)))

#SPREAD - width of the graph, or it the variance means how away is the each point from the mean 
#square root of the variance or spread is std deviation 
# Standard-Deviation - what is the average deviation of each point from the mean value 
#If standard deviation is small then spread is small 
print('\nStd-dev')
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_versicolor['petal_length']))
print(np.std(iris_virginica['petal_length']))

#we didnt do any plot here but with these value we can say that 
#mean of setosa petal-length is very less as compared to means of other two species 
#mean of versicolor and virginica is very close to each other
```
